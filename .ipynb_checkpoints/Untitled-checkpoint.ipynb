```python
import pandas as pd
import pylab as plt
import numpy as np
import matplotlib

%matplotlib inline
file = pd.read_csv("consolidado.csv", skiprows = 0, sep = ';', header=0)
file = file[file['country']=='US']
estado = pd.get_dummies(file['state'])
file = file.join(estado)

cantidad = file.groupby('creator').agg(
        {'id':'count','successful':'sum'}
   )
cantidad.columns = [ 'Cantidad','Exito']
cantidad.reset_index(inplace=True)
cantidad.columns = ['Creador', 'Cantidad','Exito']
cantidad = cantidad[cantidad['Exito']>0]
cantidad['%exito']=cantidad['Exito']/cantidad['Cantidad']
cantidad.sort_values(by = ['%exito'], inplace = True)
cantidad.reset_index(inplace=True, drop = True)
cantidad.groupby(['Cantidad']).mean()['%exito'].plot(kind = 'bar', color = 'c')
plt.ylabel("%Exito")
plt.title("% Exito vs Cantidad de Proyectos")
```

```python
#ciudades con mayor numero de patrocinadores
cantidad = file.groupby('location').agg(
        {'backers_count':'sum', 'location':'count'}
   )
cantidad.columns = ['Patrocinadores', 'Cantidad']
cantidad.reset_index(inplace=True)
cantidad.columns = ['Ciudad', 'Patrocinadores', 'Cantidad']
cantidad.sort_values(by = ['Patrocinadores'], inplace = True)
cantidad.reset_index(inplace=True, drop = True)
cantidad.tail(5).groupby(['Ciudad']).mean().plot(kind = 'barh', color = 'g')
plt.title("Ciudades Mas Patrocinadores")
plt.xlabel("Patrocinadores")
plt.ylabel("Ciudad")
plt.show()
```

```python
#ciudades con mayor numero de patrocinadores
cantidad = file.groupby('location').agg(
        {'backers_count':'sum', 'location':'count'}
   )
cantidad.columns = ['Patrocinadores', 'Cantidad']
cantidad.reset_index(inplace=True)
cantidad.columns = ['Ciudad', 'Patrocinadores', 'Cantidad']
cantidad.sort_values(by = ['Patrocinadores'], inplace = True)
cantidad.reset_index(inplace=True, drop = True)
cantidad.head(5).groupby(['Ciudad']).mean().plot(kind = 'barh', color = 'y')
plt.title("Ciudades con Menos Patrocinadores")
plt.xlabel("Patrocinadores")
plt.ylabel("Ciudad")
```
